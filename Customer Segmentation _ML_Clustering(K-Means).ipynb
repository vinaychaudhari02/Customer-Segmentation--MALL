```python
df1= data[["CustomerID","Genre","Age","Annual Income (k$)","Spending Score (1-100)"]]
X = df1[["Annual Income (k$)", "Spending Score (1-100)"]] 
from sklearn.cluster import KMeans
```

```python
wss=[]

for i in range(1,11):
    km =KMeans(n_clusters=i)
    km.fit(X)
    wss.append(km.inertia_)
print(wss)
```

```python
#elbow curve
plt.figure(figsize=(10,6))
plt.plot(range(1,11),wss)
plt.plot(range(1,11),wss, linewidth =2, color = "red", marker ="8")
plt.xlabel("KMeans")
plt.ylabel("wss")
plt.xticks(np.arange(1,11,1))
plt.show()

```

```python
km = KMeans(n_clusters = 5)
km.fit(X)
```

```python
y = km.fit_predict(X)
```

```python
df1["label"] =y
df1
```

```python
#scatter cluster plot
plt.figure(figsize =(10,6))
sns.scatterplot(x="Annual Income (k$)",y = "Spending Score (1-100)",hue = "label", palette = ['green','orange','brown','dodgerblue','red'],legend = "full",data = df1)
plt.xlabel("Annual Income")
plt.ylabel("Spendiing Score")
plt.show()
```

3D Projection

```python
df2= data[["CustomerID","Genre","Age","Annual Income (k$)","Spending Score (1-100)"]]
```

```python
X2 = df2[["Age","Annual Income (k$)","Spending Score (1-100)"]]
X2
```

```python
ls=[]

for k in range(1,11):
    km2 =KMeans(n_clusters=k,init="k-means++")
    km2.fit(X2)
    ls.append(km2.inertia_)
print(ls)
```

```python
plt.figure(figsize=(12,6)) 
plt.plot(range(1,11),ls, linewidth=2, color="red", marker ="8")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.ylabel("ls")
plt.show()
```

```python
km2 =KMeans(n_clusters=5)
km2.fit(X2)
y2 = km2.fit_predict(X2)
df2["label"] =y2
df2
```

```python
#3D Plot as we did the clustering on the basis of 3 input features
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df2.Age[df2.label==0],df2["Annual Income (k$)"][df2.label==0],df2["Spending Score (1-100)"][df2.label==0],c="purple")
ax.scatter(df2.Age[df2.label==1],df2["Annual Income (k$)"][df2.label==1],df2["Spending Score (1-100)"][df2.label==1],c="red")
ax.scatter(df2.Age[df2.label==2],df2["Annual Income (k$)"][df2.label==2],df2["Spending Score (1-100)"][df2.label==2],c="blue")
ax.scatter(df2.Age[df2.label==3],df2["Annual Income (k$)"][df2.label==3],df2["Spending Score (1-100)"][df2.label==3],c="green")
ax.scatter(df2.Age[df2.label==4],df2["Annual Income (k$)"][df2.label==4],df2["Spending Score (1-100)"][df2.label==4],c="yellow")

ax.view_init(35, 185)
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
ax.set_zlabel('Spending Score (1-100)')

plt.show()
```
